## 4. Display first 1000 rows of tables (College_A_HS, College_A_SE, College_A_SJ, College_B_HS, College_B_SE, College_B_SJ) with Python.

```python
import mysql.connector
collegedata=mysql.connector.connect(host="localhost",user="root",password="1806",database="alumni")
collegecursor=collegedata.cursor()
collegecursor.execute("select * from college_a_hs limit 1000")
happy=collegecursor.fetchall()
print("The fist 1000 rows table college_a_hs : ")
print("-"*100)
for i in happy:
    print(i)
    print("-"*100)
```
